    1) Are there certain destination or arrival cities that are home to more delays or cancellations?

    2) What are the preferred times for flights to occur?

## Dataset Overview

> This dataset reports flights in the United States, including carriers, arrival and departure delays, and reasons for delays, from 1987 to 2008.
>The data comes originally from RITA where it is described in detail.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('2008.csv')
df_airports = pd.read_csv('airports.csv')
code_city = pd.Series(df_airports.city.values,index=df_airports.iata).to_dict()
df['Origin'] = df['Origin'].replace(code_city)
df['Dest'] = df['Dest'].replace(code_city)
dict_month = {1:'January',2:'February',3:'March',4:'April'}
dict_day = {1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday',7:'Sunday'}
df['Month'] = df['Month'].replace(dict_month)
df['DayOfWeek'] = df['DayOfWeek'].replace(dict_day)
df.drop_duplicates(inplace = True)
df = df[['Month','DayOfWeek','ArrDelay','DepDelay','Distance','Cancelled','Origin','Dest']].copy()
df['ArrDelay']=df['ArrDelay'].replace(np.nan,df['ArrDelay'].mean())
df['DepDelay']=df['DepDelay'].replace(np.nan,df['DepDelay'].mean())
df.dropna(inplace=True)
```

## Origin Cities that are homes to more Delayed Flights

- The most 10 origin cities that were home to more delayed deperature flights were in order

    Pellston, Hancock, Crescent City    
    Key West, South Bend, King Salmon      
    Chicago, Mosinee, Gunnison 


- The most 10 origin cities that were home to more delayed arrival flights were in order

    Pellston, Crescent City, Hancock             
    Gillette, Key West, Gunnison          
    Champaign/Urbana, Springfield, South Bend          

```python
mean_dep_delay = df.groupby('Origin')['DepDelay'].mean().sort_values(ascending=False)[0:9]
mean_arr_delay = df.groupby('Origin')['ArrDelay'].mean().sort_values(ascending=False)[0:9]
base_color = sb.color_palette()[0]
plt.figure(figsize=[10,11])
plt.subplot(2,1,1)
sb.barplot(y=mean_dep_delay.index, x=mean_dep_delay.values, color=base_color);
plt.title('Top 10 origin cities that were home to more delayed Departure flights')
plt.xlabel('Average Departure Delay (minutes)')
plt.ylabel('City')

plt.subplot(2,1,2)
sb.barplot(y=mean_arr_delay.index, x=mean_arr_delay.values, color=base_color);
plt.title('Top 10 origin cities that were home to more delayed arrival flights')
plt.xlabel('Average Arrival Delay (minutes)')
plt.ylabel('City');
```

## Destination Cities that are homes to more Delayed Flights

- The most 10 distination cities that were home to more delayed deperature flights were in order

    Key West, Champaign/Urbana, Mosinee                  
    Newark, Dillingham, Wilkes-Barre/Scranton    
    Chicago, Hancock, Harrisburg 

- The most 10 destination cities that were home to more delayed arrival flights were in order

    Champaign/Urbana, Key West, Mosinee
    Dillingham, Newark, Chicago             
    Salem, Traverse City, San Francisco          

```python
mean_dep_delay = df.groupby('Dest')['DepDelay'].mean().sort_values(ascending=False)[0:9]
mean_arr_delay = df.groupby('Dest')['ArrDelay'].mean().sort_values(ascending=False)[0:9]

plt.figure(figsize=[10,11])
plt.subplot(2,1,1)
sb.barplot(y=mean_dep_delay.index, x=mean_dep_delay.values, color=base_color);
plt.title('Top 10 Destination cities that were home to more delayed Departure flights')
plt.xlabel('Average Departure Delay (minutes)')
plt.ylabel('City')

plt.subplot(2,1,2)
sb.barplot(y=mean_arr_delay.index, x=mean_arr_delay.values, color=base_color);
plt.title('Top 10 Destination cities that were home to more delayed Arrival flights')
plt.xlabel('Average Arrival Delay (minutes)')
plt.ylabel('City');
```

## Origin and Destination Cities that are homes to more Cancelled Flights

- The most 10 origin cities that were home to more cancelled flights flights were in order

    Pellston, Waterloo, Telluride      
    Hancock, Rhinelander, Kalamazoo      
    Hailey, Aspen, Dubuque        


- The most 10 destination cities that were home to more  cancelled flights were in order

    Pellston, Waterloo, Hancock        
    Telluride, Rhinelander, Kalamazoo
    Dubuque, Hailey, Saginaw                 

```python
origin_cancelled = df.groupby('Origin')['Cancelled'].mean().sort_values(ascending=False)[0:10]
dest_cancelled = df.groupby('Dest')['Cancelled'].mean().sort_values(ascending=False)[0:10]

plt.figure(figsize=[10,11])

plt.subplot(2,1,1)
sb.barplot(y=origin_cancelled.index, x=origin_cancelled.values*100, color=base_color);
plt.title('Top 10 Origin cities that were home to more cancelled flights')
plt.xlabel('Percentage of cancelled flights (%)')
plt.ylabel('City');

plt.subplot(2,1,2)
sb.barplot(y=dest_cancelled.index, x=dest_cancelled.values*100, color=base_color);
plt.title('Top 10 Destination cities that were home to more cancelled flights')
plt.xlabel('Percentage of cancelled flights (%)')
plt.ylabel('City');
```

## Preferred Months for Flights to occur

- February has recorded the highest average delay nearly 14 minutes delay.
- while April was the least with 8 minutes delay.

```python
plt.figure(figsize=[15,4])
plt.subplot(1,2,1)

sb.pointplot(data=df, x='Month', y='DepDelay', color=base_color);
plt.title('Departure delays across months')
plt.xlabel('Month')
plt.ylabel('Average departure Delay (minutes)');

plt.subplot(1,2,2)

sb.pointplot(data=df, x='Month', y='ArrDelay', color=base_color);
plt.title('Arrival delays across months')
plt.xlabel('Month')
plt.ylabel('Average Arrival Delay (minutes)');
```

## Preferred Days for Flights to occur

- Friday has recorded the highest average delay nearly 15 minutes.
- while wednesday was the least less than 10 minutes delay.

```python
plt.figure(figsize=[15,4])
plt.subplot(1,2,1)
sb.pointplot(data=df, x='DayOfWeek', y='DepDelay', color=base_color);
plt.title('Departure delays across Days')
plt.xlabel('Day')
plt.ylabel('Average departure Delay (minutes)');

plt.subplot(1,2,2)
sb.pointplot(data=df, x='DayOfWeek', y='ArrDelay', color=base_color);
plt.title('Arrival delays across Days')
plt.xlabel('Day')
plt.ylabel('Average arrival Delay (minutes)');
```

## Cancelled Trips associated with each month

- Februray comes with The highest cancel percentage nearly 3.5% of flights were cancelled.
- April comes with the least Percentage nearly 2% of flights were cancelled.

```python
cancelled_month = df.groupby('Month')['Cancelled'].mean().sort_values(ascending=False)
sb.barplot(x=cancelled_month.index, y=cancelled_month.values*100, color=base_color);
plt.title('Flights cancel percentage across months')
plt.xlabel('Month')
plt.ylabel('Cancelled percentage (%)');
```

## Cancelled Trips associated with each Day

- Tuesday and Friday comes with The highest cancel percentage nearly 3.2% of flights were cancelled.
- Sunday comes with the least Percentage nearly 2% of flights were cancelled.

```python
day_cancelled = df.groupby('DayOfWeek')['Cancelled'].mean().sort_values(ascending=False)
plt.figure(figsize=[7,5])
sb.barplot(x=day_cancelled.index, y=day_cancelled.values*100, color=base_color);
plt.title('Flights cancel percentage across days')
plt.xlabel('Day')
plt.ylabel('Cancelled percentage (%)');
```
