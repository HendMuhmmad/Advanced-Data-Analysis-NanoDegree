### Researching Q2: What is the most popular genre for each of the last five years?
>To get the  that most popular genre for each year we <ul><li>***First:*** get the entries of the last five years movies with average rating above or equal 7.</li></ul>

```python
df.dropna(inplace=True)
year_genre = df[(df['vote_average'] >= 7) & (df['release_year'] >= 2010)].copy()
```

> <ul> <li> Then: get a subset of this data frame to get the entries that have multiple gennres in it.</li></ul>

```python
year_genre_multiple = year_genre[year_genre['genres'].apply(lambda x :"|" in x)]
year_genre_sliced = year_genre_multiple.copy()
```

> <li> After that: split these genres as one for each entry.</li>

```python
year_genre_sliced['genres'] = year_genre_sliced['genres'].apply(lambda x : x.split("|"))
genre_list=[]

for i,sample in year_genre_sliced.iterrows():
        for genre in sample['genres']:
            item = sample.copy()
            item['genres'] = genre
            genre_list.append(item)
```

> <li>When: we finish spliting we append these value to our year_genre dataframe and drop the indexes with mulitple genres from the year_genre dataframe.</li>

```python
single_genre = pd.DataFrame(genre_list)
year_genre.drop(year_genre_multiple.index,inplace=True)
year_genre = year_genre.append(single_genre,ignore_index=True)
```

> <ul><li> Finally: we group our final dataframe according to the year, sum the number of movies with high rating and plot for each year.</li></ul>

```python
year_genre = pd.DataFrame(year_genre.groupby(['release_year'])['genres'].value_counts())
year_genre.rename(columns={'genres':'number_of_movies'}, inplace=True)
for year,new_df in year_genre.groupby(level=0):
    new_df.plot(kind='bar')
```

### Researching Q3: What is the production company that achieved the most revenue in 2015 ?
>To get the company that achieved the most revenue we:<li>***First:*** get the entries of 2015 only.</li>

```python
prod_company_2015 = df[df['release_year'] == 2015].copy()
```

><li>Then: get a subset of this data frame to get the entries that have multiple production companies in it.</li>

```python
prod_company_2015_multiple = prod_company_2015[prod_company_2015['production_companies'].apply(lambda x :"|" in x)]
prod_company_2015_sliced = prod_company_2015_multiple.copy()
```

> <li>After that: split these companies as one for each entry.</li>

```python
prod_company_2015_sliced['production_companies'] = prod_company_2015_sliced['production_companies'].apply(lambda x : x.split("|"))
company_list=[]
for i,sample in prod_company_2015_sliced.iterrows():
        for prod in sample['production_companies']:
            item = sample.copy()
            item['production_companies'] = prod
            company_list.append(item)
```

><li>When: we finish spliting we append these value to our 2015 dataframe and drop the indexes with mulitple companies from the 2015 dataframe.</li>

```python
single_company = pd.DataFrame(company_list)
prod_company_2015.drop(prod_company_2015_multiple.index,inplace=True)
prod_company_2015 = prod_company_2015.append(single_company,ignore_index=True)
```

><li>Finally: we group our final dataframe according to the company's name sum it revenue and plot the top 10 companies.</li>

```python
prod_company_2015.groupby(['production_companies'])['revenue_adj'].sum().sort_values(ascending = False)[0:10].plot(kind='bar')
```
